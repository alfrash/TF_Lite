```python
import tensorflow as tf

# store data for x and y
x = [-1, 0, 1, 2, 3, 4]
y = [-3, -1, 1, 3, 5, 7]

# create a simple keras model
model = tf.keras.models.Sequential([
    tf.keras.layers.Dense(units=1, input_shape=[1])
])

model.compile(optimizer='sgd',loss='mean_squared_error')
model.fit(x, y, epochs=500)
```

* 1- save the model in tensorflow saved file formate. 
* 2-  converte it to TFLite model.
* 3- save it to afile with extention .tflite.

```python
import pathlib

# Export the savedModel
export_dir = './tmp/saved_model'
tf.saved_model.save(model, export_dir)

# convert the model to TFLite model
converter = tf.lite.TFLiteConverter.from_saved_model(export_dir)
tflite_model = converter.convert()

# save the model
tflit_model_file = pathlib.Path('./tmp/foo.tflite')
tflit_model_file.write_bytes(tflite_model)
```
